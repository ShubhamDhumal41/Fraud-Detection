```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import itertools
from collections import Counter
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_curve, ConfusionMatrixDisplay
```

```python
# Perform Scaling
scaler = StandardScaler()
df["NormalizedAmount"] = scaler.fit_transform(df["amount"].values.reshape(-1, 1))
df.drop(["amount"], inplace= True, axis= 1)
X = df.drop(["isFraud"], axis= 1)
Y = df["isFraud"]
```

```python
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y, test_size= 0.3, random_state= 42)

print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
# DECISION TREE
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred_dt = decision_tree.predict(X_test)
decision_tree_score = decision_tree.score(X_test, Y_test) * 100
print("Decision Tree Score: ", decision_tree_score)
```

```python
# RANDOM FOREST

random_forest = RandomForestClassifier(n_estimators= 100)
random_forest.fit(X_train, Y_train)
Y_pred_rf = random_forest.predict(X_test)
random_forest_score = random_forest.score(X_test, Y_test) * 100
print("Random Forest Score: ", random_forest_score)
```

```python
# key terms of Confusion Matrix - DT

print("TP,FP,TN,FN - Decision Tree")
tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred_dt).ravel()
print(f'True Positives: {tp}')
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')
```

```python
# confusion matrix - DT

confusion_matrix_dt = confusion_matrix(Y_test, Y_pred_dt.round())
print("Confusion Matrix - Decision Tree")
print(confusion_matrix_dt,)
```

```python
# visualising confusion matrix - DT

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_dt)
disp.plot()
plt.title('Confusion Matrix - DT')
plt.show()
```

```python
# key terms of Confusion Matrix - RF

print("TP,FP,TN,FN - Random Forest")
tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred_rf).ravel()
print(f'True Positives: {tp}')
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')
```

```python
# confusion matrix - RF

confusion_matrix_rf = confusion_matrix(Y_test, Y_pred_rf.round())
print("Confusion Matrix - Random Forest")
print(confusion_matrix_rf)
```

```python
# visualising confusion matrix - RF
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_rf)
disp.plot()
plt.title('Confusion Matrix - RF')
plt.show()
```

```python
# classification report - DT

classification_report_dt = classification_report(Y_test, Y_pred_dt)
print("Classification Report - Decision Tree")
print(classification_report_dt)

# classification report - RF

classification_report_rf = classification_report(Y_test, Y_pred_rf)
print("Classification Report - Random Forest")
print(classification_report_rf)
```

```python
# ROC Curve DT

fpr, tpr, threshold = roc_curve(Y_test, Y_pred_dt)
roc_auc = metrics.auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve DT')
plt.legend(loc='lower right')
plt.show()
```

```python
# ROC Curve RF

fpr, tpr, threshold = roc_curve(Y_test, Y_pred_rf)
roc_auc = metrics.auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve RF')
plt.legend(loc='lower right')
plt.show()
```

```python
# Precision-Recall Curve DT
precision, recall, _ = precision_recall_curve(Y_test, Y_pred_dt)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve DT')
plt.legend(loc='lower left')
plt.show()
```

## Conclusion:

In the course of developing and evaluating a fraud detection model, several key observations and decisions have been made.

### Model Performance:

Both Random Forest and Decision Tree models were assessed for their performance in detecting fraudulent transactions. While the accuracy of both models was equal, it's crucial to prioritize precision in fraud detection models. Precision ensures that when a transaction is flagged as fraudulent, it's highly likely to be a true fraud, minimizing the risk of falsely accusing innocent customers. In this context, Random Forest showed higher precision, making it a preferred choice.

### Choice of Algorithms:

Random Forest and Decision Tree algorithms were selected due to their suitability for handling highly unbalanced datasets, which is common in fraud detection (99.87% legitimate transactions vs. 0.13% fraudulent transactions). Random Forest, in particular, creates multiple decision trees, simplifying the data's understanding while maintaining a balance between accuracy and precision.

### Key Factors for Fraud Prediction:

Several key factors were identified as important for predicting fraudulent customers:

1.Source of Request Security: Determining whether the source of a transaction request is secure or not is critical. Secure sources are less likely to be associated with fraudulent activity.

2.Legitimacy of the Requesting Organization: Verifying the legitimacy of the organization requesting a transaction is crucial. Illegitimate organizations are more likely to engage in fraudulent transactions.

3.Transaction History of Vendors: Analyzing the transaction history of vendors can provide insights into their legitimacy. Frequent associations with fraudulent transactions may indicate a higher risk.

### Prevention Strategies During Infrastructure Updates:
To enhance fraud prevention as the company updates its infrastructure, the following strategies are recommended:

1.Use Verified Apps Only: Encourage customers to use only verified and trusted applications to conduct financial transactions, reducing the risk of interacting with fraudulent apps.

2.Secure Browsing: Emphasize the importance of secure browsing habits. Customers should ensure they are visiting legitimate and secure websites for financial transactions.

3.Secure Internet Connections (VPN): Promote the use of Virtual Private Networks (VPNs) to secure internet connections, especially when accessing financial services remotely.

4.Device Security: Advise customers to keep their mobile devices and laptops updated with the latest security patches and antivirus software to prevent malware attacks.

5.Avoid Unsolicited Communications: Educate customers not to respond to unsolicited calls, SMS, or emails requesting sensitive financial information.

6.Immediate Contact with the Bank: If customers suspect that their security has been compromised or they have been tricked, encourage them to contact the bank immediately to report the issue and take necessary actions.

### Monitoring the Effectiveness of Prevention Measures:
To determine the effectiveness of these prevention measures, a proactive approach is essential:

1.E-Statements: Banks should actively send e-statements to customers, providing a detailed record of their account activity. Customers should review these statements regularly for any unauthorized transactions.

2.Customer Vigilance: Encourage customers to be vigilant and keep an eye on their account activity. Any unusual or suspicious transactions should be reported to the bank promptly.

3.Transaction Logs: Banks and customers should maintain transaction logs to track payments and detect any discrepancies.

#### By implementing these measures and actively involving both banks and customers in fraud prevention, the company can enhance its security infrastructure and reduce the risk of fraudulent activities. Continuous monitoring and adaptation of strategies will be essential to stay ahead of evolving fraud tactics.
